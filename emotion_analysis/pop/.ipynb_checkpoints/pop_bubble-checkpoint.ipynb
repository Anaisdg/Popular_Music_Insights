```python
import plotly.plotly as py
from plotly.grid_objs import Grid, Column
from plotly.tools import FigureFactory as FF 

import pandas as pd
import numpy as np
import time
```

```python
combined = pd.read_csv("pop_combined.csv")
combined = combined.drop(combined.columns[0], axis=1)
combined
```

```python
colors = ['#e6194b', '#3cb44b', '#ffe119', '#0082c8', '#f58231','#911eb4','#46f0f0','#f032e6','#d2f53c','#fabebe','#008080','#e6beff','#aa6e28','#fffac8','#800000','#aaffc3','#808000','#ffd8b1','#000080','#808080','#000000', '#B06A3F', '#F84EB8', '#074D26', '#586185', '#A0C1B1', '#ECF190', '#662584', '##700C05' ])
```

```python
df = pd.melt(combined, 
            id_vars='Emotion', 
            value_vars=list(combined.columns[1:]), # list of days of the week
            var_name='Years', 
            value_name='Percentile')
df.head()
```

```python
result = df.sort_values(['Emotion', "Years"], ascending=True)
result.head()
```

```python
position = pd.DataFrame(np.arange(0, len(combined), 1))
position = position.rename(columns={position.columns.values[0]:"Postion"})
position["Emotion"] = df["Emotion"]
position
```

```python
result["Size"]=result["Percentile"]
result["Genre"]="Pop"
result = result.merge(position, on="Emotion")
dataset = result
dataset
```
