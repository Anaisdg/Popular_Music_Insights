<h1><strong> Data Cleaning For Bubble Plot </strong></h1>

```python
import pandas as pd
import numpy as np
import time
```

```python
combined = pd.read_csv("rnb_combined.csv")
combined = combined.drop(combined.columns[0], axis=1)
combined
```

```python
colors = {'color':['#e6194b', '#3cb44b', '#ffe119', '#0082c8', '#f58231','#911eb4','#46f0f0','#f032e6','#d2f53c','#fabebe','#008080','#e6beff','#aa6e28','#fffac8','#800000','#aaffc3','#808000','#ffd8b1','#000080','#808080','#000000', '#B06A3F', '#F84EB8', '#074D26', '#586185', '#A0C1B1', '#ECF190', '##9ACBEA', '#662584', '##700C05']} 
len(colors)
colors_df = pd.DataFrame(colors)
colors_df['Emotion'] = combined['Emotion']
colors_df.head()
```

```python
df = pd.melt(combined, 
            id_vars='Emotion', 
            value_vars=list(combined.columns[1:]), # list of the years
            var_name='years', 
            value_name='percentile',
            )
df.head()
```

```python
result = df.sort_values(['Emotion', "years"], ascending=True)
result.head()
```

```python
position = pd.DataFrame(np.arange(1, len(combined)+1, 1))
position = position.rename(columns={position.columns.values[0]:"postion"})
position["Emotion"] = df["Emotion"]
position
```

```python
result["size"]=result["percentile"]
result["genre"]="Pop"
result = result.merge(position, on="Emotion")
result = result.merge(colors_df, on = "Emotion")
dataset = result
dataset
```

```python
dataset.rename(columns = {'Emotion': 'emotion', 'years':'year'})
```

```python
dataset.to_csv("rnb_bub_form.csv")
```
